```python
import matplotlib.pyplot as plt

plt.figure(figsize=[10,8]) #Dimensiones en pulgadas
#Data 1
time_Data = [0, 1, 2, 3, 4, 5]
#Data 2
velocity_Data = [0, 10, 20, 30, 40, 50]

#Permite hacer un grafico en donde tiene 2 parametros, 1 eje x ,2 eje y

plt.plot(velocity_Data,time_Data,marker="D",linestyle="--",color="purple")
#maker me permite personalizar el tipo de punto y linestyle el tipo de linea
#Si tuviera mas graficos, lo mismo pltplot y sus datos
plt.title("Velocity vs time")
#Title del grafico
plt.xlabel("Velocity")
#Title eje x
plt.ylabel("Time")
plt.legend() #Indica que datos se estan evaluando
plt.show()
```

```python

# Data 1
time_Data = [0, 1, 2, 3, 4, 5]

# Data 2
velocity_Data = [0, 10, 20, 30, 40, 50] 

fig, ax = plt.subplots()  # Crear subtrama
#En este caso al objeto se le trabajan las funciones
ax.plot(time_Data, velocity_Data, marker="D", linestyle="--", color="orange")
ax.set_xlabel("Time (Sg)")
ax.set_ylabel("Velocity")
ax.set_title("Velocity vs Time")
ax.grid(True)#Usamos una grid
plt.show()


```

```python

Data_x = [i for i in range(1, 101)]
Data_y = [j for j in range(101, 201)]
Data_x2= [k for k in range(201,301)]
Data_y2= [k for k in range(301,401)]
#La cantidad de datos debe de ser igual

fig, map = plt.subplots(figsize=(10,10))  # Crear la figura y los ejes

map.plot(Data_x, Data_y, color="green", linestyle="--", label="Numbers", marker="+")

map.plot(Data_x2,Data_y2,color="red",marker="^")
#Segunda linea de datos

map.set_xlabel("Numbers 1-100")

map.set_ylabel("Numbers 100-200")
#Las funciones cambiaran si uso POO ya siendo set algunas 
map.legend()  # Agregar la leyenda al gráfico

map.set_title("Numbers!")

plt.show()

```

```python

perros = [j for j in range(5, 51, 5)]  # Cantidad de perros por comuna

gatos = [1, 3, 5, 12, 15, 16, 17, 18, 12, 22]  # Cantidad de gatos por comuna

# Crear un rango para el número de comunas (10)
comunas = range(1, len(perros) + 1)

# Crear el objeto figura y el subplot
cookie, ax = plt.subplots()

ancho_barra = 0.30

ax.bar(comunas, perros, width=ancho_barra, label='Perros') #Barra de los perros 

ax.bar([i + ancho_barra for i in comunas], gatos, width=ancho_barra, label='Gatos') #Barra de los gatos
# [i + ancho_barra for i in comunas] me permite iterar sobre la barra y hace que no se monte encima de la otra

ax.set_xlabel('Comunas')
ax.set_ylabel('Cantidad')
ax.set_title('Cantidad de Perros y Gatos por Comuna')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

chicasUdem = [100, 70, 150, 50, 23]  # Cantidad de chicas en la estación
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]

# Crear el histograma
fig, ax = plt.subplots()

# Graficar el histograma
ax.hist(dias, weights=chicasUdem, bins=len(dias), color="skyblue", alpha=0.6, label="Ocurrencia en la estación UDEM")
#Solamente recibe una variable y queda en y. Bins te mostrara la cantidad de contenedores de comparacion para los datos los dias de semana (5)
#Weights le da el peso a la comparacion de frecuencia en el histograma a evaluar
#Alpha le indica la opacidad
ax.set_xlabel("Días de la semana")
ax.set_ylabel("Cantidad de chicas")
ax.set_title("Ocurrencia de chicas en la estación UDEM")
plt.style.use("ggplot")
ax.legend()

plt.show()
```

```python

#Creacion de diagramas de barras

stations = ["Udem", "Los alpes", "La palma", "Parque Belen", "Industriales"]

percentageStation = [20.5, 13.0, 8.3, 28.0, 30.2]

highlight = (0, 0, 0, 0, 0.1)

colors = ['salmon', 'aquamarine', 'mediumturquoise', 'deeppink', 'blueviolet']

plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(10, 8))

# Título del gráfico
ax.set_title("Ocurrencia en estaciones por %")

# Creación del gráfico de torta
wedges, texts, autotexts = ax.pie(percentageStation, explode=highlight, labels=stations, autopct="%1.1f%%",
                                  shadow=True, startangle=90,colors=colors)
# Leyenda
ax.legend(wedges, stations, title="Estaciones", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```
