Regresión Lineal
En pocas palabras:

La regresión lineal es un modelo estadístico que busca encontrar la relación entre una variable dependiente y y una o más variables independientes x. Se representa como una línea recta, donde la pendiente y el intercepto determinan la fuerza y dirección de la relación.

Teoría:

La ecuación básica de la regresión lineal es:



$y = β₀ + β₁x + ε$

$y= Mx+b$ Esta es otra alternativa que existe


- y: Variable dependiente
- β₀: Intercepto (valor de y cuando x es 0)
- β₁: Pendiente (cambio en y por cada unidad de cambio en x)
- E representa el término de error, que es la diferencia entre el valor observado de 
Y
y y el valor predicho por el modelo de regresión.

Para estimar los valores de β₀ y β₁, se utiliza el método de los mínimos cuadrados. Este método busca minimizar la suma de los cuadrados de las diferencias entre los valores reales de y y los valores predichos por la línea de regresión.

Interpretación:

La pendiente β₁ indica la fuerza de la relación entre x e y. Un valor mayor de β₁ indica una relación más fuerte.
El intercepto β₀ indica el valor de y cuando x es 0.
El coeficiente de determinación R² indica la cantidad de variabilidad de y que es explicada por la variable x.

# Formula Matematica de una regresion lineal



```python
modelo=LinearRegression()
```

# Entrenamos el modelo de regresion lineal

```python
modelo.fit(x,y)
```

```python
modelo.coef_ #Se obtiene el valor B_1 en la formula y = b_0 + b_1x + E PENDIENTE
```

```python
modelo.intercept_  #Se obtiene el valor B_0 en la formula y = b_0 + b_1x + E INTERCEPTO
```

```python
x
```

```python
modelo.predict([[482]]) #Predice como se compotara el valor en Y si x es 482
```

```python
plt.figure(figsize=(10,6))

#Datos sin regresion

plt.subplot(1, 2, 1) 

plt.plot(x,y,color="yellow",marker=">",label="Equivalencia autos y trafico")

plt.xlabel("Intensidad en el trafico (Autos) ")

plt.ylabel("Tiempo gastado en trafico (Minutos)")

plt.legend()

plt.grid(True)


#Regresion lineal

plt.subplot(1,2,2)

plt.plot(x,modelo.predict(x),color="purple",marker="*",label="Regresion Lineal") #Se introduce variable independiente para obtener la depediente (la opuesta)

plt.xlabel("Intensidad en el trafico (Autos) ")

plt.ylabel("Tiempo gastado en trafico (Minutos)")

plt.legend()

plt.grid(True)


plt.show()
```
