Eager Execution en TensorFlow

Eager Execution es una característica de TensorFlow que permite la evaluación de operaciones de inmediato, sin necesidad de construir gráficos computacionales previamente. En contraste con el modo "Graph Execution", donde primero se define un grafo de operaciones y luego se ejecuta, en Eager Execution cada operación se ejecuta paso a paso, lo que facilita la depuración y la experimentación interactiva.

# Broadcasting entre tensores

El "broadcasting" de tensores es un mecanismo que permite realizar operaciones aritméticas entre tensores con diferentes números de dimensiones o formas. En términos simples, el broadcasting expande automáticamente los tensores de menor tamaño para que coincidan con el tamaño del tensor más grande, permitiendo así realizar operaciones elemento a elemento¹²³⁴⁵.

Dos tensores son "transmisibles" si se cumplen las siguientes reglas¹:
- Cada tensor tiene al menos una dimensión.
- Al iterar sobre los tamaños de las dimensiones, comenzando por la dimensión final, los tamaños de las dimensiones deben ser iguales, uno de ellos es 1, o uno de ellos no existe.

Por ejemplo, si tienes dos tensores `x` e `y` con formas `(5, 3, 4, 1)` y `(3, 1, 1)` respectivamente, estos son transmisibles porque¹:
- 1ª dimensión final: ambos tienen tamaño 1.
- 2ª dimensión final: `y` tiene tamaño 1.
- 3ª dimensión final: el tamaño de `x` es igual al tamaño de `y`.
- 4ª dimensión final: la dimensión de `y` no existe.

Si los tensores `x` e `y` son "transmisibles", el tamaño del tensor resultante se calcula de la siguiente manera¹:
- Si el número de dimensiones de `x` e `y` no son iguales, se añade 1 a las dimensiones del tensor con menos dimensiones para hacerlas de igual longitud.
- Luego, para cada tamaño de dimensión, el tamaño de dimensión resultante es el máximo de los tamaños de `x` e `y` a lo largo de esa dimensión.

**Las reglas del broadcasting en operaciones con tensores**

1. **Regla de igualdad o uno**: Dos dimensiones son compatibles para el broadcasting si son iguales o si una de ellas es 1.

2. **Regla de las dimensiones faltantes**: Si los dos tensores no tienen el mismo número de dimensiones, el tensor con menos dimensiones se "expande" añadiendo dimensiones de tamaño 1 al principio hasta que ambos tensores tengan el mismo número de dimensiones.

3. **Regla de la expansión**: Si una dimensión en un tensor tiene un tamaño de 1 y la dimensión correspondiente en el otro tensor tiene un tamaño mayor que 1, el primer tensor se "expande" a lo largo de esa dimensión replicando sus valores hasta que coincida con el tamaño de la dimensión en el otro tensor.

Estas reglas se aplican de derecha a izquierda, es decir, comenzando por la última dimensión (la más a la derecha) y avanzando hacia la primera dimensión (la más a la izquierda).
 

 **Ejemplos de las reglas**

1. **Regla de igualdad o uno**: Supongamos que tienes dos tensores, `x` de forma `(3, 4)` e `y` de forma `(3, 1)`. En este caso, la primera dimensión de ambos tensores es 3, por lo que son iguales. La segunda dimensión de `x` es 4 y la de `y` es 1, por lo que también son compatibles. Por lo tanto, `x` e `y` son transmisibles.

2. **Regla de las dimensiones faltantes**: Supongamos que tienes dos tensores, `x` de forma `(5, 3, 4)` e `y` de forma `(3, 4)`. `y` tiene menos dimensiones que `x`, por lo que podemos "añadir" una dimensión de tamaño 1 al principio de `y`, haciendo que su forma sea `(1, 3, 4)`. Ahora, `x` e `y` tienen el mismo número de dimensiones y cada dimensión es compatible, por lo que `x` e `y` son transmisibles.

3. **Regla de la expansión**: Supongamos que tienes dos tensores, `x` de forma `(5, 3, 4)` e `y` de forma `(1, 3, 1)`. La primera y tercera dimensiones de `y` son 1, por lo que `y` se puede "expandir" a lo largo de estas dimensiones para coincidir con las dimensiones correspondientes de `x`. Esto significa que `y` se replica a lo largo de estas dimensiones, haciendo que su forma sea `(5, 3, 4)`. Ahora, `x` e `y` tienen la misma forma y son transmisibles.


**METODO PARA VERIFICAR SI SON TRANSMISIBLES:** Imagina que tienes dos tensores, `x` e `y`. Para determinar si estos tensores son "transmisibles", necesitas comparar las dimensiones de `x` e `y` desde la última dimensión hacia la primera. Aquí están las reglas que debes seguir:

1. Comienza desde la última dimensión (la más a la derecha si estás escribiendo las formas de los tensores). Compara el tamaño de esta dimensión en ambos tensores.
2. Si el tamaño de la dimensión en ambos tensores es el mismo, o si uno de ellos es 1, entonces esa dimensión es "transmisible".
3. Si uno de los tensores no tiene esa dimensión (porque tiene menos dimensiones), también se considera "transmisible".
4. Repite este proceso para todas las dimensiones, moviéndote de derecha a izquierda.

Si todas las dimensiones son "transmisibles" según estas reglas, entonces los dos tensores son "transmisibles" en su conjunto.

Por ejemplo, si `x` tiene forma `(5, 3, 4, 1)` y `y` tiene forma `(3, 1, 1)`, son transmisibles porque:
- La última dimensión de ambos es 1.
- La segunda última dimensión de `x` es 4 y la de `y` es 1, por lo que son transmisibles porque uno de ellos es 1.
- La tercera última dimensión de `x` es 3 y la de `y` es 3, por lo que son transmisibles porque son iguales.
- `y` no tiene una cuarta dimensión, por lo que se considera transmisible.



## Regla del 1 en los tensores SOLO TRANSMISIBLES

si un tensor tiene menos dimensiones que el otro, se considera que las dimensiones faltantes tienen un tamaño de 1. Esto es lo que se conoce como "añadir una dimensión" o "expandir una dimensión" Solamente cuando y ase sabe que son transmisibles.

Por ejemplo, si tienes un tensor `x` de forma `(5, 3, 4, 1)` y un tensor `y` de forma `(3, 1, 1)`, puedes ver que `y` tiene menos dimensiones que `x`. Para hacerlos transmisibles, puedes considerar que `y` tiene una dimensión adicional de tamaño 1 al principio, por lo que su forma se convierte en `(1, 3, 1, 1)`. Ahora, `x` e `y` tienen el mismo número de dimensiones y cada dimensión es transmisible según las reglas que mencioné antes.

Es importante tener en cuenta que este proceso de "añadir una dimensión" es conceptual y no cambia el tensor original en la memoria. Es simplemente una forma de hacer que los tensores de diferentes formas sean compatibles para ciertas operaciones. Espero que esto aclare tu duda. Si tienes más preguntas, no dudes en preguntar.

**Sin embargo, hay que tener en cuenta que esto solo funciona si las dimensiones existentes de los tensores son compatibles según las reglas de broadcasting.**

Por ejemplo, si tienes un tensor `x` de forma `(5, 3)` y un tensor `y` de forma `(5,)`, puedes considerar que `y` tiene una dimensión adicional de tamaño 1 al final, por lo que su forma se convierte en `(5, 1)`. Ahora, `x` e `y` son transmisibles porque cada dimensión es compatible: la primera dimensión es la misma (5) y la segunda dimensión de `y` es 1.

Pero si tienes un tensor `x` de forma `(5, 3)` y un tensor `y` de forma `(4,)`, no puedes hacerlos transmisibles simplemente añadiendo una dimensión a `y`. Esto se debe a que la primera dimensión de `x` y `y` no es compatible: 5 no es igual a 4 y ninguno de ellos es 1.



```python
# Definimos algunos tensores
a = tf.constant([1, 2, 3])  # Tensor de forma (3,)
b = tf.constant([[0], [1], [2]])  # Tensor de forma (3, 1)

# Broadcasting en la operación de suma
c = a + b  # El resultado es un tensor de forma (3, 3)

# También podemos hacer broadcasting con la función tf.add
d = tf.add(a, b)  # El resultado es un tensor de forma (3, 3)

# Imprimimos los resultados
print("c = \n", c.numpy())
print("d = \n", d.numpy())
```

# Ragged Tensor

Los "Ragged Tensors" son una característica de TensorFlow que permite manejar datos con formas no uniformes¹². En otras palabras, son la equivalencia en TensorFlow de listas anidadas de longitud variable.

Aquí hay algunos ejemplos de datos que podrían ser representados como Ragged Tensors¹²:

- Características de longitud variable, como el conjunto de actores en una película.
- Lotes de entradas secuenciales de longitud variable, como frases o clips de video.
- Entradas jerárquicas, como documentos de texto que se subdividen en secciones, párrafos, oraciones y palabras.


```python
rt = tf.ragged.constant([[1, 2, 3], [], [4, 5], [6], []])

print(rt)
```

```python
rt.shape
```



El tensor `rt` que creamos es un RaggedTensor, lo que significa que puede tener un número variable de elementos a lo largo de algún eje. En este caso, `rt` tiene 5 elementos en su primer eje (o "filas"), y cada uno de estos elementos es una lista de longitud variable.



Como se evidencia , el primer elemento (o "fila") de `rt` es la lista `[1, 2, 3]`, que tiene 3 elementos. El segundo elemento de `rt` es la lista `[]`, que no tiene elementos. El tercer elemento de `rt` es la lista `[4, 5]`, que tiene 2 elementos. Y así sucesivamente.


# Sparse Tensor

Un "Sparse Tensor" o tensor disperso es una estructura de datos que se utiliza para representar tensores que contienen principalmente valores cero¹⁴. En lugar de almacenar todos los valores, un tensor disperso solo almacena los valores que son diferentes de cero, junto con sus ubicaciones¹⁴. Esto puede ser muy eficiente en términos de memoria y computación cuando se trabaja con datos muy dispersos¹.

TensorFlow representa un tensor disperso como tres tensores densos separados: `indices`, `values` y `dense_shape`¹⁴:

- `indices`: Un tensor 2D con forma `[N, rank]`, que contiene los índices de los valores no cero.
- `values`: Un tensor 1D con forma `[N]` que contiene todos los valores no cero.
- `dense_shape`: Un tensor 1D con forma `[rank]`, que especifica la forma del tensor.

Estos tres componentes se recogen en un objeto `tf.sparse.SparseTensor` para facilitar su uso¹⁴.

```python

# Crear un SparseTensor
st = tf.sparse.SparseTensor(indices=[[0, 3], [2, 4]], values=[10, 20], dense_shape=[3, 10])

print(st)
```

```python
# We can convert sparse tensors to dense

print(tf.sparse.to_dense(st))
```
