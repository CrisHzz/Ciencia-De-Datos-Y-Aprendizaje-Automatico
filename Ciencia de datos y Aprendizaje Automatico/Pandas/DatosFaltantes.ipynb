# Eliminacion de filas 

```python
titanic.shape
```

```python
titanic_remake=titanic.dropna(axis=0,how="all") #Con esto eliminamos filas en donde todos sus valores son faltantes


titanic_remake.shape

#Se evidencia que no hay filas con esta especialidad entonces no se puede usar asi
```

```python
titanic_remake2=titanic.dropna(axis=0,how="any") #Se eliminan las filas que tiene al menos un valor faltante

titanic_remake2.shape
```

# Rellenar datos faltantes 

```python
#Rellenando datos de la edad

pd.isnull(titanic["age"]).values.ravel().sum() #Me retorna la cantidad de datos en la columna age que son faltantes
```

```python
columnas_faltantes=titanic["age"].fillna(titanic["age"].mean()) #Se rellenan los datos faltantes de age con su promedio

columnas_faltantes.head(20)
```

```python
titanic_remake["new_age"]=columnas_faltantes #Insertar la nueva columna en titanic_remake

titanic_remake.head(50)
```

```python
titanic_remake=titanic_remake[["survived","name","sex","new_age","cabin","boat","home.dest"]]

titanic_remake.head(50) #Se elimino la tabla age con datos faltantes y ya esa columna no tiene problemas
```

```python
titanic_remake.columns.values

titanic_final=titanic_remake #Crea una nueva tabla


titanic_final["boat"].fillna("Unknown", inplace=True) #Reemplaza donde hay un NaN por un Unknown


titanic_final["boat_mean"] = titanic_final["boat"] #Remplaza la nueva tabla por esa

# Select relevant columns
titanic_final = titanic_final[["survived", "name", "sex", "new_age", "cabin", "boat_mean", "home.dest"]] #Crea la tabla con las nuevas columnas


```
