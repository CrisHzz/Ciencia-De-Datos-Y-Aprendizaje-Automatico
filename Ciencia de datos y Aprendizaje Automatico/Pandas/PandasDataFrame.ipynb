```python
# Forma correcta de reformar en tablas mas pequeñas con doble parantesis por ser tuplas
nacho_gordos = nacho[["short_name", "age", "weight_kg"]]

# Mostrar las primeras filas de nacho_gordos
print(nacho_gordos.head(10))
```

Obtener el dato de un jugador en especifico con varias condiciones

```python
# Obtener el índice del jugador más gordo
indice_papa_gordo = nacho_gordos["weight_kg"].idxmax()

# Obtener la información completa del jugador más gordo
papa_gordo_info = nacho_gordos.loc[indice_papa_gordo] #tambien le puedo añadir para ser sofisticado las columnas ["short_name", "age", "weight_kg"]

# Imprimir la información
print("El jugador más gordo es:")

print(papa_gordo_info)
```

nacho_gordos: Es el DataFrame que contiene la información de los jugadores, específicamente las columnas "short_name" (nombre corto), "age" (edad) y "weight_kg" (peso en kilogramos).

.loc[]: Es un método de indexación en pandas que se utiliza para seleccionar filas y columnas por etiquetas. En este caso, se está utilizando para seleccionar una fila específica (índice indicado por indice_papa_gordo) y ciertas columnas ("short_name", "age", "weight_kg").

indice_papa_gordo: Es el índice del jugador más gordo, obtenido utilizando el método idxmax() en la columna "weight_kg".

["short_name", "age", "weight_kg"]: Es una lista de nombres de columnas que estamos seleccionando para obtener información específica del jugador más gordo. En este caso, estamos interesados en el nombre corto, la edad y el peso en kilogramos.

```python
indice_papa_gordo_minimo=nacho_gordos['weight_kg'].idxmin()

info_papa_gordo_minimo=nacho_gordos.loc[indice_papa_gordo_minimo,["short_name", "age", "weight_kg"]]

print(f"El mas gordo de nacional es {info_papa_gordo_minimo}")
```

```python
jovenes_nacho=nacho.sort_values(by="age",ascending=True) #Ordenar por orden ascedente una tabla, si es False es ascendente

print(jovenes_nacho)
```

```python
mas_gordo_joven_id = nacho[(nacho['weight_kg'] == nacho['weight_kg'].idxmax()) & (nacho['age'] == nacho['age'].idxmin)].index

mas_gordo_joven_data = nacho.loc[mas_gordo_joven_id, ["short_name", "age", "weight_kg"]]

print(f"El jugador más gordo y joven es:\n{mas_gordo_joven_data}")

#No hay jugador joven y mas gordo
```
